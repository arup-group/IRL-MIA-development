```python
import os
import geopandas as gpd
import pandas as pd

# Base directory containing the 'shp' folder
BASE_DIR = r"C:\Users\alden.summerville\Documents\dev-local\IRL-MIA-development\jupyter\outputs\shp"

# List of date folder names to process
date_folders = ["2025-02-05_1134_1", 
                "2025-02-05_1148_2", 
                "2025-02-05_1253_3",
                "2025-02-05_1308_4",
                "2025-02-05_1338_5",
                "2025-02-05_1341_6",
                ]

# Determine the latest timestamp to use for naming
latest_timestamp = max(date_folders)  # Full timestamp format YYYY-MM-DD_HHMM

# Create output folder name with timestamp
output_folder_name = f"{latest_timestamp}_Combined"
output_folder_path = os.path.join(BASE_DIR, output_folder_name)

# Ensure the output folder exists
os.makedirs(output_folder_path, exist_ok=True)

# Define output shapefile name with timestamp
output_shapefile = os.path.join(output_folder_path, f"Microwatersheds_Combined_{latest_timestamp}.shp")

# Initialize an empty list to hold GeoDataFrames
gdfs = []

# Iterate through the given date folders and read the GeoPackage
for folder in date_folders:
    folder_path = os.path.join(BASE_DIR, folder)
    
    if not os.path.isdir(folder_path):
        print(f"Skipping {folder_path} (not a valid directory)")
        continue
    
    # Find the .gpkg file in the folder
    gpkg_files = [f for f in os.listdir(folder_path) if f.endswith(".gpkg")]
    
    if not gpkg_files:
        print(f"No GeoPackage found in {folder_path}")
        continue
    
    # Assuming only one .gpkg per folder
    gpkg_path = os.path.join(folder_path, gpkg_files[0])
    
    # Read the GeoPackage
    gdf = gpd.read_file(gpkg_path)
    gdfs.append(gdf)

# Concatenate all GeoDataFrames if any were found
if gdfs:
    combined_gdf = gpd.GeoDataFrame(pd.concat(gdfs, ignore_index=True))
    
    # Ensure the CRS is set (assuming all files have the same CRS)
    combined_gdf.set_crs(gdfs[0].crs, inplace=True)
    
    # Write to a shapefile
    combined_gdf.to_file(output_shapefile, driver="ESRI Shapefile")
    
    print(f"Combined shapefile saved to: {output_shapefile}")
else:
    print("No valid geopackages found to combine.")
```
