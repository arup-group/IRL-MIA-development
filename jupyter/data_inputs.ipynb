```python
# Libraries
from pysheds.grid import Grid
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
# File paths

# DEM
dem_path = fr'data-inputs\\Pineda_Scalgo_8m_NAD83.tif'
# 

# NHD Flowlines
flowlines = gpd.read_file(r'data-inputs\\NHD-Flowlines\\FLA-NHD-Flowlines_NAD83.shp')

# Shoreline
shoreline = gpd.read_file(r'data-inputs\\Shoreline\\FLA-Shoreline_4269.shp')

# Ponds
ponds = gpd.read_file(r'data-inputs\\IRL-Ponds-Export\\IRL-Ponds-Export_4269.shp')

# Nutrients
nutrients = gpd.read_file(r'data-inputs\\Nutrients\\Nutrients_Brevard_4326.shp')

# Impervious area
impervious_raster = r'data-inputs\\ImperviousArea\\FlaImperviousArea_4326.tif'

# Land Cover
land_cover = gpd.read_file(r'data-inputs\\LandCover\\Land_Cover_FLA_4326.shp')
```

```python
# Plot the DEM

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
import seaborn as sns

grid = Grid.from_raster(dem_path)
dem = grid.read_raster(dem_path)

# Define the normalization for the elevation between 0 and 15 meters
norm = colors.Normalize(vmin=0, vmax=15)

fig, ax = plt.subplots(figsize=(8,6))
fig.patch.set_alpha(0)

# Plot the DEM with normalization applied
plt.imshow(dem, extent=grid.extent, cmap='terrain', norm=norm, zorder=1)
#plt.imshow(dem, extent=grid.extent, cmap='terrain', zorder=1)
plt.colorbar(label='Elevation (m)')
plt.grid(zorder=0)
plt.title('Digital elevation map', size=14)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.tight_layout()
```
